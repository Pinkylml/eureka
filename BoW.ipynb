```python
import re
import os
import nltk
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
import string
from collections import defaultdict
```

```python
#REad the stop words list
stop_words_paht='data/reuters/stopwords'
stop_words=[]
with open(stop_words_paht,'r') as file:
    content=file.read().split('\n')
    stop_words.append(content)
    
print(stop_words)
```

```python
def normalize_text(text):
    text = text.lower()
    #text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

```python
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
data_directory = 'data/reuters/training/'
filenames=os.listdir(data_directory)
filenames.sort(key=lambda x: int(x))
print(filenames)
```

```python
data_directory = 'data/reuters/training/'
word_counts = defaultdict(int)
wset = set()
nltk.download('stopwords')
stop_words2 = set(stopwords.words('english'))
stem_docs = []


for filename in filenames:
    with open(os.path.join(data_directory, filename), 'r') as file:
        content = file.read()         
        content = normalize_text(content) 
        sdoc = []
        for word in content.lower().translate(str.maketrans('', '', string.punctuation)).split(" "):
            if word not in stop_words[0] or word not in stop_words2:
                #word=stemmer.stem(word)
                word=lemmatizer.lemmatize(word)
                wset.add(word)
                word_counts[word] += 1 
                sdoc.append(lemmatizer.lemmatize(word))
        stem_docs.append(sdoc)                                                    
```

```python
sorted_word_count = sorted(word_counts.items(), key=lambda item: item[1], reverse=True)
print(sorted_word_count)
print(len(sorted_word_count))
```

```python
print(len(stem_docs))
print(len(filenames))
```

```python
ev_docs = []
for i, doc in enumerate(stem_docs):
    ev_doc = []
    print("Procesando archivo: ",filenames[i],"...")
    for dim in sorted_word_count:
        if dim[0] in doc:
            ev_doc.append(1)
        else:
            ev_doc.append(0)
    ev_docs.append(ev_doc)
    print("Archivo procesado: ",filenames[i]," OK")
print(ev_docs)
```
